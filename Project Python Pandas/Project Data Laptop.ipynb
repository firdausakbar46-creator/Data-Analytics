# Exploratory Data Analysis

# Company

```python
grafik_1 = sns.countplot(x= 'Company', data = df2)

for bar in grafik_1.containers:
    grafik_1.bar_label(bar)

plt.title("Company Count")
plt.xticks(rotation= 65)
plt.show()
```

Berdasarkan grafik diatas menunjukkan bahwa perusahaan yang paling banyak kuantitasnya adalah Dell daan Lenovo

# Operation System

```python
grafik_2 = sns.countplot(x= 'Operation System', data= df2, hue= 'Operation System')

for bar in grafik_2.containers:
    grafik_2.bar_label(bar)

plt.title("Count of Operation System")
plt.xticks(rotation= 65)
plt.show()
```

Berdasarkan grafik diatas terlihat bahwa Windows 10 adalah sistem operasi yang palingg banyak diproduksi

# Type Name

```python
sns.set(rc= {'figure.figsize':(20,5)})
grafik_3 = sns.countplot(data= df2, x= 'Type Name', hue= 'Type Name')

for bar in grafik_3.containers:
    grafik_3.bar_label(bar)

plt.title("Count of Type Name")
plt.show()
```

Berdasarkan grafik diatas menunjukkan bahwa Notebook adalah tipe nama yang kuantitasnya paling banyak
