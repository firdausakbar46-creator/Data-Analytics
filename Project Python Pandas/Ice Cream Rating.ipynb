# Project Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Asus\Downloads\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('bmh')
```

```python
df.plot(figsize = (7,5), title = 'Ice Cream Rating', ylabel = 'Scores', xlabel = 'Daily Rating')
```

```python
df.plot(kind = 'bar' ,stacked = True, figsize = (7,5), ylabel = 'Scores', xlabel = 'Daily Rating')
```

```python
df['Texture Rating'].plot(kind = 'bar', stacked = True, figsize = (7,5))
```

```python
df.plot.barh(stacked = True, figsize = (7,5))
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 300, figsize= (7,5), color = 'Red')
```

```python
df.plot.hist(bins = 20, figsize = (7,5))
```

```python
df.boxplot(figsize = (7,5))
```

```python
df.plot.area(figsize = (7,5))
```

```python
df.plot.pie(y = 'Overall Rating', figsize = (7,5))
```
