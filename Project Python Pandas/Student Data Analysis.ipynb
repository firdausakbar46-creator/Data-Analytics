# Exploratory Data Analaysis

### Gender

```python
# membuat grafik jumlah gender

grafik_gender = sns.countplot(x = 'sex', data = df2, hue= 'sex')

for bar in grafik_gender.containers:
    grafik_gender.bar_label(bar)


plt.title('Grafik Jumlah Gender')
plt.show()
```

Dari grafik diatas terlihat bahwa kuantitas perempuan lebih banyak dari laki-laki

### Pekerjaan Ibu

```python
grafik_mjob = sns.countplot(x= 'Mother Job', data= df2, hue = 'Mother Job')

for bar in grafik_mjob.containers:
    grafik_mjob.bar_label(bar)

plt.title('Grafik Jumlah Pekerjaan Ibu')
plt.show()
```

Berdasarkan grafik diatas terlihat bahwa kategori pekerjaan ibu 'lainnya' paling tinggi, sedangkan paling tinggi kedua adalah 'services'

### Pekerjaan Ayah

```python
grafik_fjob  = sns.countplot(x = 'Father Job', data= df2, hue = 'Father Job')

for bar in grafik_fjob.containers:
    grafik_fjob.bar_label(bar)


plt.title('Grafik Pekerjaan Ayah')
plt.show()
```

Berdasarkan grafik diatas terlihat bahwa kategori pekerjaan ayah 'lainnya' paling tinggi, sedangkan paling tinggi kedua adalah 'services'
