# Exploratory Data Analysis

### Count of Top 10 University by Country

```python
df_head = df.head(10)
df_head
```

```python
grafik_1 = sns.countplot(x= 'Country/Territory', data = df_head, hue= 'Country/Territory')

for bars in grafik_1.containers:
    grafik_1.bar_label(bars)

sns.set(rc={'figure.figsize':(10,5)})
plt.title('Jumlah Top 10 Universitas Berdasarkan Negara')
plt.show()
```

Berdasarkan grafik diatas terlihat bahwa jumlah universitas top 10 paling banyak berada di negara 'United States of America' dan 'United Kingdom'

### Count of University by Status

```python
grafik_2 = sns.countplot(x= 'Status', data= df, hue= 'Status')

for bars in grafik_2.containers:
    grafik_2.bar_label(bars)

sns.set(rc={'figure.figsize':(10,5)})
plt.title('Jumlah Top 10 Universitas Berdasarkan Status')
plt.show()
```

Berdasarkan grafik diatas terlihat bahwa Top 10 Universitas paling banyak berstatus 'public'

### Count of University by Size

```python
grafik_3 = sns.countplot(x= 'Size', data= df, hue= 'Size')

for bars in grafik_3.containers:
    grafik_3.bar_label(bars)

sns.set(rc= {'figure.figsize':(10,5)})
plt.title('Jumlah Universitas Berdasarkan Ukuran')
plt.show()
```

Berdasarkan grafik diatas terlihat bahwa ukuran universitas yang paling banyak adalah 'L'
