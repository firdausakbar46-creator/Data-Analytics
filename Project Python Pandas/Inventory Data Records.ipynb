```python
df2 = df.rename(columns = {"Product ID":"ID Produk", "Product Name":"Nama Produk" ,"Opening \nStock":"Pembukaan Stok Barang", "Purchase/\nStock in":"Pembelian Stok Barang", "Number of \nUnits Sold":"Jumlah Unit Terjual", "Cost Price \nPer Unit (USD)":"Biaya Barang Per Unit", "Cost Price\nTotal (USD)":"Total Biaya Barang"})
df2
```

```python
df2.set_index("Nama Produk")
```

```python
pivot1 = df2.pivot_table(index= "Nama Produk", values = "Jumlah Unit Terjual")
pivot1
```

```python
pivot1.plot.bar(title = "Grafik Jumlah Unit yang Terjual",figsize=(10,7))
```
